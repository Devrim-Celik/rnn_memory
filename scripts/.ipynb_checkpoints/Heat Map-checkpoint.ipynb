```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[0.46286875009536743, 0.678543746471405,  0.8339999914169312, 0.8846250772476196, 0.95478755235672],
                 [0.6992250084877014,  0.9630374908447266, 0.9784625172615051, 0.9861937761306763, 0.9764062762260437],
                 [0.8577375411987305,  0.8310750126838684, 0.8218749761581421, 0.9679562449455261, 0.9672562479972839],
                 [0.7280812859535217,  0.9194687604904175, 0.8673499822616577, 0.9029437899589539, 0.9074000120162964],
                 [0.7682499289512634,  0.8463062047958374, 0.9079750180244446, 0.8958936929702759, 0.8349499702453613]])

mean = np.mean(data)
std = np.std(data)

normalized_data = (data - mean)/std
```

```python
plt.figure(figsize = (8,8))
res = plt.imshow(normalized_data, cmap='hot', interpolation='nearest', extent=[0.85, 1.25, 0.75, 0.25])
plt.savefig("heatmap.png")
plt.show()
```
