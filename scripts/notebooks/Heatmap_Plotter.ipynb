```python
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import os

font = {'size'   : 26}

matplotlib.rc('font', **font)
```

```python
prefix = "heatmap_"
plots = {
    'ar_auto':(26,14),
    'ar_fw':(20,18),
    'mnist_fw':(20,15),
    'mnist_auto':(20,4),
    'mnist_auto_2':(20,4)
}
ar_auto = "ar_auto"
ar_fw = "ar_fw"
mnist_fw = "mnist_fw"
mnist_auto_1 = "mnist_auto"
mnist_auto_2 = "mnist_auto_2"
suffix = ".csv"

selected_file = ar_auto

filename = os.path.join("../data",prefix + selected_file + suffix)
```

```python
df = pd.read_csv(filename, sep=";")

headers = df.columns.values.tolist()
ylab = headers[0]
xlab = headers[1]

plt.figure(figsize = plots[selected_file])
sns.set(font_scale=4)
df = df.pivot(ylab, xlab, "value")
#df[xlab] = df[xlab].astype(str)
ax = sns.heatmap(df)
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.savefig("../plots/"+prefix+selected_file+".png", bbox_inches='tight', pad_inches=1)
plt.show()
```
